```python
img = np.zeros((600,600,3),np.uint8)
cv2.namedWindow("Bars")
cv2.createTrackbar("on","Bars",0,255,wscube)
```

```python
while True:
    cv2.imshow("Bars",img)
    if cv2.waitKey(1) & 0xff == ord("q"):
        break
    
    on = cv2.getTrackbarPos("on","Bars")
    img[:] = [0,0,on]
cv2.destroyAllWindows()
```
