1. White Blank Image

```python
new_img1 = np.ones((700,600,3),np.uint8)*255
```

```python
print(new_img1.shape)
print(new_img1)
```

2. Black Blank Image

```python
new_img2 = np.zeros((700,600,3),np.uint8)*255
```

```python
print(new_img2.shape)
print(new_img2)
```

```python
cv2.imshow("White Blank Image",new_img1)
cv2.imshow("Black Blank Image",new_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
