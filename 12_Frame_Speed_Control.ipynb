# Slow And Fast Motion video

a video is basically a set of moving pictures moving at a rate of about 24 frames per second. If this frame rate is decreased then we get a slow-motion video and that is what we are intended to do. To capture a video we need to create a VideoCapture object. Then we will create a VideoWriter object for writing the output slow-motion video.

```python
import cv2
```

```python
cap = cv2.VideoCapture(r"path\video.mp4")

while cap.isOpened():
    rat, frame = cap.read()
    if not rat:
        print("frame is empty")
        break
    frame = cv2.resize(frame,(700,600))
    cv2.imshow("speed control testing",frame)
    if cv2.waitKey(6) & 0xff == ord("q"):
        break
    
cap.release()
cv2.destroyAllWindows()
    
```
