```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\Screenshot43.png")
img = cv2.resize(img ,(500,600))
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
kernal = np.ones((5,5),np.int8)
print(kernal)
kernal
```

3. Opening 

```python
# morphologyEx(src, op, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]])

opening = cv2.morphologyEx(img , cv2.MORPH_OPEN , kernal , iterations =1)
```

4. Closing

```python
# 

closing = cv2.morphologyEx(img , cv2.MORPH_CLOSE , kernal ,iterations =1)
```

5. Gradient

```python
gradient = cv2.morphologyEx(img , cv2.MORPH_GRADIENT , kernal , iterations = 1)
```

6. Black Hat

```python
black_hat = cv2.morphologyEx(img , cv2.MORPH_BLACKHAT , kernal , iterations =1)
```

7. Top Hat

```python
top_hat =  cv2.morphologyEx(img , cv2.MORPH_TOPHAT , kernal , iterations =1)
```

```python
cv2.imshow("gray image",img)
cv2.imshow("opening image",opening)
cv2.imshow("Closing image",closing)
cv2.imshow("Gradient image",gradient)
cv2.imshow("Black hat",black_hat)
cv2.imshow("Top Hat",top_hat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
