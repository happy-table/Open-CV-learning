# Border in opencv

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\Screenshot 2026-02-01 213727.png")
img_resize = cv2.resize(img ,(600,500))
```

```python
top = 20
bottom =20
left =20
right =20
borderType = cv2.BORDER_CONSTANT
dst = None
value = 2

img1 = cv2.copyMakeBorder(img_resize, top, bottom, left, right, borderType, dst, value)
img1_resize = cv2.resize(img1,(600,500))
```

```python
h = np.hstack((img_resize,img1_resize))
```

```python
cv2.imshow("bordering image",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
