# Object Tracking And Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(r"opencv_Images\Background_Subtraction_Tutorial_frame.mp4")
r,f = cap.read()
x,y,w,h = 255, 217, 30, 95  # roi

t = (x,y,w,h)
roi = f[y:y+h,x:x+w]

hsv_roi = cv2.cvtColor(roi,cv2.COLOR_BGR2HSV)
# cv2.imshow("hsv roi",hsv_roi)

mask = cv2.inRange(hsv_roi ,np.array((0.,60.,32.)), np.array((180.,255.,255.)))

roi_histogram = cv2.calcHist([hsv_roi] ,[0] , mask,[180] ,[0,180])
nor = cv2.normalize(roi_histogram,roi_histogram,0,255, cv2.NORM_MINMAX)
cv2.imshow("test",roi)
# print(f.shape)

tr = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT , 10 ,1)

while cap.isOpened():
    r, f = cap.read()
    if r == True:
        # f = cv2.resize(frame,(500,400))
        hsv_f = cv2.cvtColor(f,cv2.COLOR_BGR2HSV)
        d = cv2.calcBackProject([hsv_f],[0], roi_histogram ,[0,180] ,1)

        # moving of object-tracking
        r, t = cv2.CamShift(d,t,tr)
        x,y,w,h = t
        final =cv2.rectangle(f ,(x,y),(x+w,y+h) , (0,0,255),2)
        
        cv2.imshow("tracking object",final)
        if cv2.waitKey(100) & 0xff == ord("q"):
            break
    else:
        break
        
cap.release()
cv2.destroyAllWindows()
```
