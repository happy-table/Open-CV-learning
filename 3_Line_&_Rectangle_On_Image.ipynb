# Drawing Line

```python
import cv2
```

```python
old_img = cv2.imread(r"path\id1.png")
```

```python
old_resize = cv2.resize(old_img ,(300,500))
```

```python
img = old_resize
pt1 = (20,60)
pt2 = (50,190)
color = (0,255,0)
thickness = 4

new_img = cv2.line(img,pt1,pt2,color,thickness)
```

```python
cv2.imshow("text",old_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Drawing Rectangle

```python
import cv2
```

```python
old_img = cv2.imread(r"id1.png")
```

```python
old_resize = cv2.resize(old_img ,(1000,780))
```

```python
img = old_resize
pt1 = (63,131)
pt2 = (439,430)
color = (255,0,0)
thickness = 4

new_img = cv2.rectangle(img,pt1,pt2,color,thickness)
```

```python
cv2.imshow("rectangle",old_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
