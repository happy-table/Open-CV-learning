# Backgraound Subtraction

```python
import cv2
import numpy as np
```

```python
org_v =  cv2.VideoCapture(r"path\videoPro.mp4")
sub_m = cv2.createBackgroundSubtractorMOG2()
while True:
    r, frame = org_v.read()
    if r== True:
        frame = cv2.resize(frame,(600,500))
        sub_v = sub_m.apply(frame)
        cv2.imshow("subtracted background video",sub_v)
        cv2.imshow("original video",frame)
        if cv2.waitKey(25) & 0xff==("q"):
            break
    else:
        break
org_v.release()
cv2.destroyAllWindows()
                          
```
