```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\Screenshot22.png")
img.shape
```

```python
cv2.circle(img,(160,69),4,(0,0,255),-1)
cv2.circle(img,(306,60),4,(0,0,255),-1)
cv2.circle(img,(377,233),4,(0,0,255),-1)
cv2.circle(img,(72,227),4,(0,0,255),-1)
```

```python
160,69
306,60
377,233
72,227
```

```python
w,h = (386,316)

src1 = np.float32([[138,32],[305,59],[0,210],[378,227]])
src1
dst1 = np.float32([[0,0],[w,0],[0,h],[w,h]])
m = cv2.getPerspectiveTransform(src1,dst1)
new_img = cv2.warpPerspective(img,m,(w,h))
```

```python
cv2.imshow("prospective image",new_img)
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
