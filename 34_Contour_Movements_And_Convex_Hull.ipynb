````python
# c : contour
cnt,h = cv2.findContours(th , cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
````

````python
new_img = cv2.drawContours(img , cnt ,-1 ,(80,0,160),2)
new_img = cv2.resize(new_img,(900,600))
````

````python
ar = []
````

````python
# finding the center with help of moments()
for c in cnt:
    m = cv2.moments(c)
    if m["m00"] != 0:                     # ✅ only calculate if area is not zero
        x = int(m["m10"] / m["m00"])
        y = int(m["m01"] / m["m00"])
        cv2.circle(img, (x, y), 5, (0, 0, 255), -1)   # mark center
    else:
        print("Skipped a tiny/empty contour")  # optional, for debugging

    cv2.drawContours(img, [c], -1, (255, 0, 0), 2)    # ✅ wrap c in a list []
    a = cv2.contourArea(c)
    ar.append(a)

    # epsilon
    ep = 0.1*cv2.arcLength(c,True)   # 0.1 ,0.01 , 0.001 ... -> for 
    d = cv2.approxPolyDP(c , ep ,True)

    # Convex Hull
    h = cv2.convexHull(d)
    
    # drawing rectangle on max area covered
    x,y,w,h = cv2.boundingRect(h)
    cv2.rectdangle(img ,(x,y) ,(x+w,y+h) ,(0,255,0) ,3)

print(ar)
````

````python
import cv2
import numpy as np

img = cv2.imread(r"opencv_Images\Screenshot42.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img.shape)

# Convert to binary
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for contour in contours:
    # Skip tiny contours (noise)
    if cv2.contourArea(contour) < 100:
        continue

    # Find center using moments
    m = cv2.moments(contour)
    if m["m00"] != 0:
        cx = int(m["m10"] / m["m00"])
        cy = int(m["m01"] / m["m00"])

        # Draw contour outline
        cv2.drawContours(img, [contour], -1, (0, 255, 0), 2)

        # Draw center dot
        cv2.circle(img, (cx, cy), 5, (0, 0, 255), -1)

        # Print coordinates
        print(f"Center: ({cx}, {cy})")

        # Label on image
        cv2.putText(img, f"({cx},{cy})", (cx+10, cy),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

cv2.imshow("Centers", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
import cv2
import numpy as np

img = cv2.imread(r"opencv_Images\Screenshot42.png")
img = cv2.resize(img, (500, 500))

gr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# ✅ FIX 1: Lower threshold value (127 instead of 220)
# 220 was too high — it missed most colored shapes
_, th = cv2.threshold(gr, 150, 255, cv2.THRESH_BINARY_INV)

# ✅ FIX 2: Use RETR_EXTERNAL — only outer contours, no tiny inner ones
cnt, h = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print(f"Total contours found: {len(cnt)}")

# Draw contours on a copy
new_img = img.copy()
cv2.drawContours(new_img, cnt, -1, (80, 0, 160), 2)

ar = []

# Finding center with moments
for i, c in enumerate(cnt):

    # ✅ FIX 3: Skip tiny noise contours (area too small)
    if cv2.contourArea(c) < 500:
        print(f"Skipped tiny contour {i}")
        continue

    m = cv2.moments(c)
    if m["m00"] != 0:
        x = int(m["m10"] / m["m00"])
        y = int(m["m01"] / m["m00"])

        # Draw center dot in red
        cv2.circle(new_img, (x, y), 6, (0, 0, 255), -1)

        # Draw coordinates as text
        cv2.putText(new_img, f"({x},{y})", (x + 8, y),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 0), 1)

        print(f"Shape {i+1} → Center: ({x}, {y}), Area: {int(cv2.contourArea(c))}")

    cv2.drawContours(new_img, [c], -1, (255, 0, 0), 2)

    a = cv2.contourArea(c)
    ar.append(a)

    # epsilon
    ep = 0.01*cv2.arcLength(c,True)   # 0.1 ,0.01 , 0.001 ... -> for 
    d = cv2.approxPolyDP(c , ep ,True)

    # Convex Hull
    h = cv2.convexHull(d)
    
    # drawing rectangle on max area covered
    x,y,w,h = cv2.boundingRect(h)
    cv2.rectangle(new_img ,(x,y) ,(x+w,y+h) ,(0,255,0) ,2)

new_img = cv2.resize(new_img, (900, 600))
cv2.imshow("Contour image", new_img)
cv2.imshow("Original image",img)
cv2.imshow("GrayScale image",gr)
cv2.waitKey(0)
cv2.destroyAllWindows()

# ```

# ---

# ## Why `220` Failed — Visualized
# ```
# Pixel values in your image:
#   White background  →  ~255
#   Shape edges/shadow→  ~200-230   ← your threshold 220 cut HERE
#   Solid shape color →  ~50-180

# With threshold=220 (BINARY_INV):
  # 255 → 0 (black)   ✅ background removed
#   220 → 0 (black)   ⚠️  shadow also removed
#   180 → 255 (white) ✅ shape kept
#   BUT edges become broken fragments → tiny contours → area = 0 → skipped!

# With threshold=127 (BINARY_INV):
#   255 → 0 (black)   ✅ background removed
#   127+ → 0 (black)  ✅ cleaner cut
#   <127 → 255 (white)✅ full solid shapes captured → good contours!
````
