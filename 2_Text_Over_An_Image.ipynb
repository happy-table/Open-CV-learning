# Text On Image

```python
import cv2
```

```python
img_get = cv2.imread(r"id1.png")
```

```python
resized = cv2.resize(img_get ,(600,700))
```

```python
img = resized
txt = "Child Vaccine"
origin = (50,100)
fontFace = cv2.FONT_HERSHEY_DUPLEX
fontSize = 2
color = (20,63,96)
thickness = 2
linType = cv2.LINE_8
bottomLeftOrigin = False

text = cv2.putText(img ,txt ,origin ,fontFace ,fontSize ,color , thickness ,linType ,bottomLeftOrigin )
```

```python
cv2.imshow("resized image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
