# Arithmetic Operations On Image

# Merge 2 Images

```python
import cv2
```

```python
img1 = cv2.imread(r"path\d1.jpg")
img2 = cv2.imread(r"path\d3.jpg")
```

```python
img1_resize = cv2.resize(img1,(800,500))
img2_resize = cv2.resize(img2,(800,500))
```

```python
new_img = cv2.addWeighted( img1_resize ,1,img2_resize,0.3,2)
```

```python
new_img = cv2.subtract(img1_resize ,img2_resize)
```

```python
cv2.imshow("blended images",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
