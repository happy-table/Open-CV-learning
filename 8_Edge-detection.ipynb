# Image Rotation

```python
import cv2
```

```python
img1 = cv2.imread(r"path\d1.jpg")
 
```

```python
img_resize = cv2.resize(img1,(600,600))
```

getRotationMatrix2D(center , angle , scale)

```python
w,h = img_resize.shape[0],img_resize.shape[1]

center = (w/2,h/2)
angle = 50
scale = 0.71

m = cv2.getRotationMatrix2D(center , angle , scale)
m
```

```python
# src -> source
src = img_resize
metrix = m
dsize = (h,w)

new_img = cv2.warpAffine(src ,metrix,dsize)
```

```python
cv2.imshow("rotated image",new_img )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Image Blurring

Real images have noise:
1. sensor noise
2. lighting variations
3. compression artifacts

Blurring averages nearby pixels, so random variations cancel out.

This helps a lot in:
1. edge detection
2. contour detection
3. thresholding

```python
import cv2
import numpy as np
```

```python
org_img = cv2.imread(r"path\img2.png")
```

```python
img_resize = cv2.resize(org_img,(300,400))
```

# Gaussian Blur
GaussianBlur(src , ksize , sigmaX , dst , sigmaY ,borderType)

```python
g = cv2.GaussianBlur(img_resize ,(7,7),0)
m = cv2.medianBlur(img_resize,1)
b = cv2.bilateralFilter(img_resize,5,75,54)
```

```python
h = np.hstack((img_resize,g,m,b))
```

```python
cv2.imshow("blur",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
