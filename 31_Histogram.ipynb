#  Analyze an Image using Histogram

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\Screenshot217.png")
img.shape
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blank_w = np.ones((500,500,3),np.uint8)*255
blank_b = np.zeros((500,500,3),np.uint8)*255
```

```python
hist_w = cv2.calcHist([blank_w], [0] ,None ,[256],[0,256])
hist_b = cv2.calcHist([blank_b], [0] ,None ,[256],[0,256])
```

```python
plt.subplot(2,1,1)
plt.plot(hist_w)
plt.subplot(2,1,2)
plt.plot(hist_b)
plt.show()
```

```python
plt.figure(figsize=(10,6))
hist_img = cv2.calcHist([img],[0],None ,[255],[0,255])

plt.plot(hist_img)
plt.show()
```

```python
# cv2.imshow("black window",blank_b)
# cv2.imshow("wight window",blank_w)
cv2.imshow("Original image",img)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
img2 = cv2.imread(r"opencv_Images\Screenshot25.png")
```

```python
plt.figure(figsize=(10,10))

b,g,r = cv2.split(img2)

hist1 = cv2.calcHist([b],[0],None,[256],[0,256])
hist2 = cv2.calcHist([g],[0],None,[256],[0,256])
hist3 = cv2.calcHist([r],[0],None,[256],[0,256])

plt.subplot(3,1,1)
plt.plot(hist1)
plt.subplot(3,1,2)
plt.plot(hist2)
plt.subplot(3,1,3)
plt.plot(hist3)

# plt.hist(b.ravel(),255,(0,255))
# plt.hist(g.ravel(),255,(0,255))
# plt.hist(r.ravel(),255,(0,255))

plt.show()
```

```python
plt.hist(b.ravel(),255,range = (0,255), color = "blue")
plt.hist(g.ravel(),255,range = (0,255), color = "green")
plt.hist(r.ravel(),255,range = (0,255),color = "red")

plt.show()
```
