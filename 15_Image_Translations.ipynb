```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\Screenshot37.png")
img = cv2.resize(img , (500,500))
                 
```

```python
metrix = np.float32([[4,3,100],[3,4,50]])
new_img = cv2.warpAffine(img, metrix ,(500,500))
```

```python
cv2.imshow("original image",img)
cv2.imshow("matrix",metrix)
cv2.imshow("new image",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
