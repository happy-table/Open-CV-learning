```python
import cv2
import numpy as np

# This function is required by createTrackbar
# It runs whenever the trackbar value changes
# We don't need to use it, so we keep it empty
def color_tracker(x):
    pass

# Create a blank black image
# Shape = 700x700 pixels
# 3 channels (BGR)
# uint8 means pixel values range from 0 to 255
img = np.zeros((700,700,3), np.uint8)

# Create a named window where everything will appear
cv2.namedWindow("color")

# createTrackbar(name, window_name, initial_value, max_value, callback_function)
cv2.createTrackbar("R", "color", 0, 255, color_tracker)
cv2.createTrackbar("G", "color", 0, 255, color_tracker)
cv2.createTrackbar("B", "color", 0, 255, color_tracker)

while True:
    
    # Get current positions of trackbars
    # These values change when you slide them
    r = cv2.getTrackbarPos("R", "color")
    g = cv2.getTrackbarPos("G", "color")
    b = cv2.getTrackbarPos("B", "color")

    # Fill entire image with selected color
    # Remember OpenCV uses BGR (not RGB)
    img[:] = [b, g, r]

    # Show the updated image
    cv2.imshow("color", img)

    # Wait 1 millisecond
    # If user presses 'q', exit loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Close all OpenCV windows
cv2.destroyAllWindows()
```
