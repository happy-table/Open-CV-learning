# Extract Image from Video

```python
import cv2
```

```python
cap = cv2.VideoCapture(r"path\Pro.mp4")
```

```python
c=0
while True:
    r,frame = cap.read()
    if r== True:
        frame = cv2.resize(frame,(600,500))
        filename = "org_img"+str(c)+".png"
        cv2.imwrite(filename , frame)
        c=c+1

        cv2.imshow("video",frame)
        if cv2.waitKey(25) & 0xff== ord("q"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
