# Saving frames on device

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\Screenshot 2026-02-01 212925.png")
img_resize = cv2.resize(img,(500,300))
```

```python
h = np.hstack((img_resize,img_resize))
v = np.vstack((h,h))
```

```python
cv2.imwrite("4in1_img.png",v)
```

```python
cv2.imshow("image",v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
