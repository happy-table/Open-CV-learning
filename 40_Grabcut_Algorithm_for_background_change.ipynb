```python
import cv2
import numpy as np

img = cv2.imread(r"opencv_Images\img36.1.jpg")

# Create initial mask
mask = np.zeros(img.shape[:2], np.uint8)

# Background & foreground models
bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)

# Better rectangle (adjust this manually for your image)
rect = (100, 100, 700, 500)

# Apply GrabCut with rectangle
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

# ----------- REFINEMENT STEP -----------
# Mark definite foreground (person area)
mask[200:450, 250:600] = cv2.GC_FGD

# Mark definite background
mask[:50, :] = cv2.GC_BGD
mask[:, :50] = cv2.GC_BGD

# Run GrabCut again using mask
cv2.grabCut(img, mask, None, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_MASK)

# Create final mask
mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')

# Apply mask
result = img * mask2[:, :, np.newaxis]

# Show result
cv2.imshow("Foreground", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
