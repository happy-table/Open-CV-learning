# Saving a Video

```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

f = cv2.VideoWriter_fourcc(*"XVID")
out = cv2.VideoWriter("demo1.avi",f ,20 ,(frame_width, frame_height))

while cap.isOpened():
    r , frame = cap.read()
    if r == True:
        frame = cv2.flip(frame,1)
        out.write(frame)
        cv2.imshow("video",frame)
        if cv2.waitKey(1) & 0xff == ord("q"):
            break
    else:
        break
    
cap.release()
out.release()
cv2.destroyAllWindows()
print("COMPLETED")
```

```python
cam = cv2.VideoCapture(r"open_cv2\demo.mp4")

while cam.isOpened():
    rat, frame = cam.read()
    if not rat:
        print("frame is empty")
        break
    frame = cv2.resize(frame,(500,400))
    cv2.imshow("speed control testing",frame)
    if cv2.waitKey(1) & 0xff == ord("q"):
        break
    
cam.release()
cv2.destroyAllWindows()
print("all frames displayed successfully")
```
