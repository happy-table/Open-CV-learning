# Mouse Event Bindings

```python
import cv2
import numpy as np
```

```python
def python(event ,x,y,f,p):
    if event ==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),10,(0,0,140),6)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(img,(x,y),(x+20,y+20),(150,0,0),6)
```

```python
cv2.namedWindow("wight screen named")
img = np.ones((600,600,3),np.uint8)*255
cv2.setMouseCallback("wight screen named",python)

while True:
    #cv2.imshow("wscube",img)
    cv2.imshow("wight screen named1",img)
    if cv2.waitKey(25) & 0xff ==ord("q"):
        break
cv2.destroyAllWindows()
```
