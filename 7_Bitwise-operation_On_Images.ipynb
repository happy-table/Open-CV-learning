# Operations On Binary Images

Black and white images
1. AND
2. OR
3. NOT
4. EXOR

```python
import cv2
import numpy as np
```

import images

```python
img1 = cv2.imread(r"path\img1.png")
img2 = cv2.imread(r"path\img3.png")
```

resize imported images

```python
img1_resize = cv2.resize(img1,(500,600))
img2_resize = cv2.resize(img2,(500,600))
```

bitwise AND operator

```python
new_img1 = cv2.bitwise_and(img1_resize,img2_resize)
```

bitwise OR operator

```python
new_img2 = cv2.bitwise_or(img1_resize,img2_resize)
```

bitwise NOT operator on img2

```python
new_img3 = cv2.bitwise_not(img2_resize)
```

bitwise NOT operator on img1

```python
new_img4 = cv2.bitwise_not(img1_resize)
```

bitwise XOR operator

```python
new_img5 = cv2.bitwise_xor(img1_resize,img2_resize)
```

horizontal stack of array of images and bitwise operator

```python
h = np.hstack((img1_resize,img2_resize,new_img5))
```

```python
cv2.imshow("Array of img1, img2 and binary image",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
