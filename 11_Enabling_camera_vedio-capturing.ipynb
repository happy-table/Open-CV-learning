# Enabling a Camera And Capturing frames

```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
while True:
    r , frame = cap.read()
    if r == True:
        frame = cv2.resize(frame ,(600,600))
        cv2.imshow("captures frames",frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Quitting webcam feed...")
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
