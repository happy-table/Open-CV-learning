# Histogram Equalizer

```python
1. Adaptive equalizer
2. CLAHE equaliser :Contrast Limited Adaptive Histogram Equalization
```

# 1. Adaptive Equalizer

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"opencv_Images\Screenshot532.png")
img = cv2.resize(img,(500,500))
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
hist_img = cv2.calcHist([img],[0],None,[255],[0,255])
hist_gray =cv2.calcHist([gray],[0],None,[255],[0,255])

plt.subplot(2,1,1)
plt.plot(hist_img)
plt.subplot(2,1,2)
plt.plot(hist_gray)
plt.show()
```

```python
eq = cv2.equalizeHist(gray)

hist_eq = cv2.calcHist([eq],[0],None,[255],[0,255])

plt.plot(hist_eq)
plt.show()
```

```python
cv2.imshow("Original image",img)
cv2.imshow("Gray Scale image",gray)
cv2.imshow("Equalised image",eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 2. CLAHE equalizer

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"opencv_Images\img2.png")
img = cv2.resize(img,(500,400))
img.shape

gr = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
hist_img = cv2.calcHist([img],[0],None,[255],[0,255])
hist_gr = cv2.calcHist([gr],[0],None,[255],[0,255])

plt.subplot(2,1,1)
plt.plot(hist_img,c="red")
plt.subplot(2,1,2)
plt.plot(hist_gr, c="green")
plt.show()
```

```python
cl = cv2.createCLAHE(clipLimit =1 , tileGridSize =(8,8))
f= cl.apply(gr)
```

```python
hist_cl = cv2.calcHist([f],[0],None ,[255],[0,255])
plt.plot(hist_cl)
```

```python
hist_cl = cv2.calcHist([f],[0],None ,[255],[0,255])
hist_img = cv2.calcHist([img],[0],None,[255],[0,255])
hist_gr = cv2.calcHist([gr],[0],None,[255],[0,255])

plt.plot(hist_cl,c="blue")
plt.plot(hist_img,c="red")
plt.plot(hist_gr, c="green")
```

```python
cv2.imshow("Original image",img)
cv2.imshow("GrayScale image",gr)
cv2.imshow("CLAHE equilized image",f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
