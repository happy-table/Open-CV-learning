# Video Background Removal Using Algorithm

```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(r"path\videoPro.mp4")

# using 2 algo to remove background from video

algo1 = cv2.createBackgroundSubtractorKNN(detectShadow = True)

algo2 = cv2.createBackgroundSubtractorMOG2(detectShadow = True)


while True:
    r,frame = cap.read()
    
    if r == True:
        f = cv2.resize(frame,(600,500))
        result1 = algo1.apply(f)
        result2 = algo2.apply(f)
        cv2.imshow("algo1",result1)
        cv2.imshow("algo2",result2)
        cv2.imshow("frames",f)
        if cv2.waitKey(50) & 0xff == ord("q"):
            break
    else:
        break

    
cap.release()
cv2.destroyAllWindows()
```
