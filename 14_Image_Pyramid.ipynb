# Image Pyramid

```python
1. to dec resolution and shrink the image.
2. analyse multiple images for different things.
3. the shrinking goes to harf of previous image in each case.
```

```python
import cv2
```

1. pyrDown()

```python
img = cv2.imread(r"opencv_Images\Screenshot5.png")
print(img.shape)

new = cv2.pyrDown(img)
print(new.shape)

new1 = cv2.pyrDown(new)
print(new1.shape)

new2 = cv2.pyrDown(new1)
print(new2.shape)
```

```python
cv2.imshow("image",img)
cv2.imshow("new image",new)
cv2.imshow("new 1st image",new1)
cv2.imshow("new 2nd image",new2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

2. pyrUp()

```python
big_new = cv2.pyrUp(new2)
print(big_new.shape)

big_new1 = cv2.pyrUp(big_new)
print(big_new1.shape)

big_new2 = cv2.pyrUp(big_new1)
print(big_new2.shape)
```

```python
cv2.imshow("big new image",big_new)
cv2.imshow("big new 1st image",big_new1)
cv2.imshow("big new 2nd image",big_new2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
