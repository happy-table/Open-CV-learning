# Polygon On Image

```python
import cv2
import numpy as np
```

```python
old_img = cv2.imread(r"path\id1.png")
```

```python
old_resize = cv2.resize(old_img ,(1000,780))
```

```python
img = old_resize
pts = np.array([[700,40],[600,90],[400,240],[200,500]])
isClosed = True
color = (165,178,5)
thickness = 4
lineType = 15

new_img = cv2.polylines(img , [pts] , isClosed , color , thickness,lineType)
```

```python
cv2.imshow("Plygon image",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
