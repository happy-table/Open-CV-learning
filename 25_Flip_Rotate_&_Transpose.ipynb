# Flip , Rotate & Transpose

```python
import cv2
```

```python
img = cv2.imread(r"opencv_Images\Screenshot17.png")
img = cv2.resize(img,(500,400))
```

# 1. Flipping

```python
# cv2.flip() method is used to flip a 2D array.
#  The function cv.flip() flips a 2D array around vertical, horizontal, or both axes.

horizontal_flip_img = cv2.flip(img,1)
vertical_flip_img = cv2.flip(img,0)
both_flip_img = cv2.flip(img,-1)
```

```python
cv2.imshow("image",img)
cv2.imshow("horizontal flipped image",horizontal_flip_img)
cv2.imshow("vertical flipped image",vertical_flip_img)
cv2.imshow("Both horizontal and vertically flipped image",both_flip_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 2. Rotation

```python
rotated_180_img = cv2.rotate(img,cv2.ROTATE_180)
rotated_90cw_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
rotated_90acw_img = cv2.rotate(img , cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
cv2.imshow("Original image",img)
cv2.imshow("180 rotated",rotated_180_img)
cv2.imshow("90 clockwise rotated",rotated_90cw_img)
cv2.imshow("90 anticlock wise rotated",rotated_90acw_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 3. Transpose

```python
transpose_img = cv2.transpose(img)
print(transpose_img.shape)
print(img.shape)
```

```python
cv2.imshow("Original image",img)
cv2.imshow("transpose image",transpose_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
