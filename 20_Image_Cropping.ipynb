# Cropping the Image

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\Screenshot25.png")
img = cv2.resize(img,(700,600))
```

```python
print(img.shape)
```

```python
# [y1:y2,x1:x2]
# P1 =(x1,y1)  , P2 =(x2,y2)
# -ve --> from the end
crop_img = img[370:-30,370:-30]
```

```python
cv2.imshow("resized image",img)
cv2.imshow("Cropped Image",crop_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
