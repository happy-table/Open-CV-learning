# Circle

```python
import cv2
```

```python
old_img = cv2.imread(r"path\id1.png")
```

```python
old_resize = cv2.resize(old_img ,(1000,780))
```

```python
img = old_resize
center = (380,610)
radius = 60
color = (0,180,0)
thickness = 3
lineType = 16

new_img = cv2.circle(img , center , radius , color , thickness , lineType)
```

```python
cv2.imshow("circled image",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Ellips

```python
import cv2
```

```python
old_img = cv2.imread(r"path\id1.png")
```

```python
old_resize = cv2.resize(old_img ,(1000,780))
```

```python
img = old_resize
center = (380,610)
axes = (60,100)
angle = 45
startAngle = 0
endAngle = 360
color = (0,180,0)
thickness = 3
lineType = 16

new_img = cv2.ellipse(img , center ,axes , angle , startAngle , endAngle , color , thickness , lineType)
```

```python
cv2.imshow("circled image",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
