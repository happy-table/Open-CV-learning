# Template Matching
(project --> google photo search features)

```python
import cv2
import numpy as np
```

```python
img1 = cv2.imread(r"opencv_Images\pictures\Bombeiros-de-resgatam-seis-pessoas-e-um-cachorro-que-estavam-ilhados-em-Garuva-03-scaled.jpeg")
template_img = cv2.imread(r"opencv_Images\pictures\img_search(38).jpg")
# img2 = cv2.resize(template_img,(200,200))

gry1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gry2 = cv2.cvtColor(template_img,cv2.COLOR_BGR2GRAY)

w,h = gry2.shape[1],gry2.shape[0]
print(gry2.shape)
result = cv2.matchTemplate(gry1, gry2 ,cv2.TM_CCORR_NORMED)
print(result)

thr = 0.97  # threshold

# create array
l = np.where(result >= thr)
print(l)          # array data from location

for i in zip(*l[::-1]):
    cv2.rectangle(img1,i,(i[0]+w ,i[0]+h) ,(0,0,255) ,1)
```
