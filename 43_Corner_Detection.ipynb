# Corner Detection

# 1.Harris Corner Detection

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\pictures\pessoas_seus_pets_Carly_Davidson_01.jpg")
img = cv2.resize(img,(500,500))
gr = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gr = np.float32(gr)

res = cv2.cornerHarris(gr,2,3,0.04)
res = cv2.dilate(res ,None)
print(res)
print("Max value:", res.max())
print("Min value:", res.min())

img[res>0.01*res.max()] = [0,0,255]
```

```python
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 2.Shi-Tomasi Corner Detection

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\pictures\cachorro_gato.jpg")
img = cv2.resize(img,(600,500))
gr = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

cr = cv2.goodFeaturesToTrack(gr , 80,0.02,30)

cr = np.int64(cr)
# print(cr)
for i in cr:
    x,y = i.ravel()
    cv2.circle(img,(x,y),5,(0,0,200),-1)
```

```python
cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
