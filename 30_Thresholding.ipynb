```python
import cv2
```

```python
img = cv2.imread(r"opencv_Images\Screenshot17.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(900,600))
```

```python
#_,th1 = cv2.threshold(img ,180,255,cv2.THRESH_BINARY)
#_,th2 = cv2.threshold(img,88,255,cv2.THRESH_BINARY_INV)
#_,th3 = cv2.threshold(img,88,255,cv2.THRESH_DRYRUN)
#_,th4 = cv2.threshold(img,150,255,cv2.THRESH_MASK)
_,th5 = cv2.threshold(img,50,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
#cv2.imshow("Binary threshold",th1)
#cv2.imshow("Binary inverse threshold",th2)
#cv2.imshow("Dry run threshold",th3)
#cv2.imshow("Mask threshold",th4)
cv2.imshow("Mask threshold",th5)
cv2.imshow("original image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
