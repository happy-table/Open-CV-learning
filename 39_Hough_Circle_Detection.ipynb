# Hough Circle Detection

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"opencv_Images\ChatGPT Image Mar 22, 2026, 02_20_07 AM.png")
img = cv2.resize(img,(900,700))
gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img.shape

circle = cv2.HoughCircles(gry,cv2.HOUGH_GRADIENT,1,20, param1 = 50, param2 = 80 , minRadius = 3 , maxRadius = 50)
if circle is not None:
    data = np.uint16(np.around(circle))

    for (x, y, r) in data[0, :]:
        cv2.circle(img, (x, y), r, (0, 255, 0), 1)
else:
    print("No circles detected")

```

```python
cv2.imshow("Image",img)
# cv2.imshow("grayscale image",gry)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
